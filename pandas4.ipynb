Handling missing data in pandas:

When we downloading data from an internet or any other resource, it may having missing values. so to deal with missing data we can use three different methods:

1. fillna(): to fille missing values using different ways.
2. interpolate: to make guess on missing values using interpolation.
3. dropna: to drop rows having missing values.

```python
import pandas as pd
df=pd.read_csv("weather_data2.csv")
df
```

```python
type(df.day[0])
```

```python
# So here the day column is string not date, we first want to change the column to date, to do it we use parse_dates:
df=pd.read_csv("weather_data2.csv", parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
# Now we want to set day column as index, so to do that, we use set_index method:
df=pd.read_csv("weather_data2.csv", parse_dates=["day"])
df.set_index("day", inplace=True)
df
```

```python
df.loc["2017/01/10"]
```

```python
# So now we want to set some meaningful values instead of NaN values, so the first methood is fillna():
fillna_df=df.fillna(0)
fillna_df
```

```python
# Now if you think, sometimes zero (0) value doesn't give us better sense, like in event column right here, so we have to change it for a specific column:
fillna_df=df.fillna({
    "temperature":0,
    "windspeed": 0,
    "event": 'no event'
})
fillna_df
```

```python
# Now again we're not happy with this filling, see for the temperature for the data '1/1/2017' we have 32 degree but on tomorrow it becomes zero, so when we do mean operation, it will carry out to wrong answer, so we want to fill it with some estimate values...
# One approach here is to use ffill() method, means: carry with previous value.
fillna_df=df.fillna(method="ffill")
fillna_df
```

```python
# Next approach will be 'bfill', it carry with the coming value:
fillna_df=df.fillna(method="bfill")
fillna_df
```

```python
# Next you can use 'axis', which copy beside column value, Means: it will copy data horizently:
fillna_df=df.fillna(method="bfill", axis="columns")
fillna_df
```

```python
# Limit will limit copying values, means: to how many next or previous cell it must be copied.
fillna_df=df.fillna(method="ffill", limit=1)
fillna_df
```

```python
# for more properties related to fillna(), go to pandas documentation...
```

```python
""" 
Again we're still not happy with this filling, because on Jan 1 temp is 32 and on Jan 4 is also 32 and also on Jan 5 is 28, so on Jan 4 the temperature should be between Jan 1 and Jan 5. that's what we guess. 
So to deal with this point we have a method called interpolate in pandas, using this method we can do linear interpolation:
So interpolation takes the middle value: 
"""
iplt_df=df.interpolate()
iplt_df
```

```python
# Here we missed the dates Jan 2 and Jan 3, so the temperature should be assign related to that dates, to do that we can use time method:
iplt_df=df.interpolate(method="time")
iplt_df
```

```python
# Third approach is dropping those rows which have missing values:
new_df=df.dropna()
new_df
```

```python
# Sometimes we want to drop only those rows which has all empty cells, not those which have some empty ceslls, so for that we can use 'how' parameter:
# index is not counting as cell data.
new_df=df.dropna(how="all")
new_df
```

```python
# We can also specify that which rows must be droped based on number of empty cells, for that we can use 'thresh' parameter:
new_df=df.dropna(thresh=2)
new_df
```

```python
# The last thing is, to insert the missing dates, for that 1st we define a range, then we pass it to dateTimeIndex and we re-index our dataframe:
dt=pd.date_range("01/01/2017", "01/11/2017")
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```

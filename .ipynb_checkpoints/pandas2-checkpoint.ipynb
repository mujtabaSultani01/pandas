## Pandas Tutorial_part2
**Dataframe is most commonly used object in pandas. It is a table like datastructure containing rows and columns similar to excel spreadsheet.** Today we will cover how to create a Data Frame? We have different ways of creating a Data Frame:*

1. Using CSV file
2. Using Excel File
3. From Python Dictionary
4. From list of Tuples
5. From list of Dictionaries

To know on advanced level then go and check **pandas documentation**...

```python
# 1) Using CSV file:
import pandas as pd
df = pd.read_csv("weather_data.csv")
df
```

```python
# 2) Using Excel file:
df2 = pd.read_excel("users_file.xlsx", "Sheet1")
df2.head(4)
```

```python
# 3) Using Python Dictionary:
weather_data = {
    'day': ['1/1/2018','1/2/2018','1/3/2017','1/4/2018','1/5/2018','1/6/2018'],
    'temperature': [22,25,32,29,30,27],
    'windspeed': [6,7,2,7,4,2],
    'event': ['Rain', 'Snow', 'Sunny','Sunny','Sunny', 'Snow']
}
df3 = pd.DataFrame(weather_data)
df3
```

```python
# 4) Using list of Tuples:
weather_data = {
    ('1/1/2022', 32, 6, 'Snow'),
    ('1/2/2022', 37, 4, 'Sunny'),
    ('1/3/2022', 31, 5, 'Snow'),
    ('1/4/2022', 27, 8, 'Rain'),
    ('1/5/2022', 34, 5, 'Sunny')
}
df4 = pd.DataFrame(weather_data, columns = ["day", "temperature", "windspeed", "event"])
df4
```

```python
# 5) Using list of Dictionary:
weather_data = [
    {"day":'1/1/2022', "temperature":32, "windspeed":6, "event":'Snow'},
    {"day":'1/2/2022', "temperature":37, "windspeed":4, "event":'Sunny'},
    {"day":'1/3/2022', "temperature":31, "windspeed":5, "event":'Snow'},
    {"day":'1/4/2022', "temperature":27, "windspeed":8, "event":'Rain'},
    {"day":'1/5/2022', "temperature":34, "windspeed":5, "event":'Sunny'},
]
df5 = pd.DataFrame(weather_data)
df5
```

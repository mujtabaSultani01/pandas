Handling missing data in pandas: Today we'll use replace method to deal with missing data.

```python
import pandas as pd
import numpy as np
df=pd.read_csv("weather_data1.csv")
df
```

```python
# Again here we have weather data, -99999 means the value is missing, instead of empty cells -99999 is most hapning when you dealing with messy data.
# so the first task is to replace the value of -99999 to none.
new_df = df.replace(-99999, value=np.NaN)
new_df
```

```python
# if we want to change a set of values to a specific value then:
new_df = df.replace(to_replace=[-99999,-88888], value=0)
new_df
```

```python
# We can change all the -99999 and 'event=>0' values to nane:
new_df=df.replace({
    'temperature': -99999,
    'windspeed': -99999,
    'event': "0",
}, np.NaN)
new_df
```

```python
# Now if we change the NaN values to no event in 'event' column, then using specific values we can do mapping...
new_df = df.replace({
        -99999: np.nan,
        'no event': 'Sunny',
    })
new_df
```

```python
# Now sometimes we have characters with numerical values, for example here if we have 'C' or 'F' with temperature and 'mph'
# windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True) 
new_df
```

```python
# We can replace a list of values into another list of values:
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```

```python
df.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4])
```

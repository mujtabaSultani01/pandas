**Pandas Melt Function:** Similar to pivot_table, melt function is used to transform or reshape data.

Let's see, we have the following data which has weather data of three different cities...

Now for data analysis purpose you may change the format of the data suitable for your analysis

![](weatherByCities.png)

You want your data to be in the followin format:

![](weatherByCities1.png)

```python
import pandas as pd
df=pd.read_csv("weather4.csv")
df
```

```python
# So for that we can use melt simply, id_vars is the thing you want in the X-axis.
df1=pd.melt(df, id_vars=["day"])
df1
```

```python
# Once you have this dataset, we can do bunch of things like filter on top of it:
# filter could be useful when you have a big dataset.
df1=pd.melt(df, id_vars=["day"])
df1[df1["variable"]=="chicago"]
```

```python
# Here in the dataset we have 'variable' name for cities column, to change the variable name:
df1=pd.melt(df, id_vars=["day"], var_name="city")
df1
```

```python
# To change the temperature column name, we can simply use value_name:
df1=pd.melt(df, id_vars=["day"], var_name="city", value_name="temperature")
df1
```

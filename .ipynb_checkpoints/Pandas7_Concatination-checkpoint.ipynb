## Pandas Tutorial_part7
**Pandas Concatinate:** Concatinate is an operation when you want to join two or more DataFrames.

```python
import pandas as pd
# First DataFrame
india_weather = pd.DataFrame({
    "city": ["Mumbai", "Delhi", "Banglor"],
    "temperature": [34, 45, 30],
    "humidity": [80, 60, 78], 
})
india_weather
```

```python
# Second DataFrame
us_weather=pd.DataFrame({
    "city": ["new yark", "chicago", "orlando"],
    "temperature": [21, 14, 35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
# DataFrames concatinating...
df=pd.concat([india_weather, us_weather])
df
```

```python
# To get a continue index, we can use ignore_index parameter:
df=pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
# To check for more parameters of 'concat' function, check Pandas-Concat documentation.
# We can also create a sub index for each sub DataFrame:
df=pd.concat([india_weather, us_weather], keys=["india", "us"])
df
```

```python
# We can use this indexi as loc() fun:
df.loc["india"]
# The usage of this indexes is, when we have big concatinated DataFrames and want to retrieve sub DataFrame.
```

```python
# Till now we opened two DataFrames on top of each other, but sometimes we want to add values to the side...
temperature_df=pd.DataFrame({
    "city": ["mumbai", "delhi", "banglor"],
    "temperature": [32, 45, 30]
})
temperature_df
```

```python
# Next DataFrame:
windspeed_df=pd.DataFrame({
    "city": ["mumbai", "delhi", "banglor"],
    "windspeed": [7, 12, 9]
})
windspeed_df
```

```python
df=pd.concat([temperature_df, windspeed_df])
df
```

```python
# Now we can add 'windspeed' column to the original DataFrame(df):
df=pd.concat([temperature_df, windspeed_df], axis=1)
df
```

```python
# Now see if we have some miss city or values, and also change order:
windspeed_df=pd.DataFrame({
    "city": ["delhi", "mumbai" ],
    "windspeed": [7, 12]
})
windspeed_df
```

```python
# So the result will be look like this:
df=pd.concat([temperature_df, windspeed_df], axis=1)
df
```

```python
# As we know they go just by rows and add the values. so this is look to be incorrect, in order to fix the problem we can use index.
# in phase of DataFrame creation we can pass index:
temperature_df=pd.DataFrame({
    "city": ["mumbai", "delhi", "banglor"],
    "temperature": [32, 45, 30],
}, index=[0, 1, 2]
)
temperature_df
```

```python
# For 'windspeed' DataFrame also:
windspeed_df=pd.DataFrame({
    "city": ["delhi", "mumbai" ],
    "windspeed": [7, 12],
}, index=[1, 0])
windspeed_df
```

```python
# Now we can join them in correct way using index:
df=pd.concat([temperature_df, windspeed_df], axis=1)
df
```

```python
# We can also append 'Series' to the DataFrame:
S=pd.pandas.Series(["Humid", "Dry", "Rain"], name="event")
S
```

```python
df=pd.concat([temperature_df, S], axis=1)
df
```
